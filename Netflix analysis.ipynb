```python
import pandas as pd

# Hum khud se Netflix ka ek sample data bana rahe hain
data = {
    'show_id': ['s1', 's2', 's3', 's4', 's5'],
    'type': ['Movie', 'TV Show', 'Movie', 'TV Show', 'Movie'],
    'title': ['3%', '9', '23:59', 'Transformers', 'Narutos'],
    'director': ['Jorge Michel', 'Shane Acker', 'Gilbert Chan', 'Michael Bay', 'Kishi'],
    'country': ['Brazil', 'United States', 'Singapore', 'United States', 'Japan'],
    'release_year': [2020, 2009, 2011, 2015, 2021],
    'rating': ['TV-MA', 'PG-13', 'R', 'PG-13', 'TV-14']
}

# Is data ko table (Dataframe) mein badal rahe hain
df = pd.DataFrame(data)

# Data dekhne ke liye
df.head()
```

```python
# Yeh code count karega ke Movies aur TV Shows kitni hain
df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Movies aur TV Shows ka Bar Chart banana
df['type'].value_counts().plot(kind='bar', color=['darkred', 'grey'])

# Chart ke upar headings aur labels lagana
plt.title('Netflix Content Type')
plt.xlabel('Type')
plt.ylabel('Count')

# Chart ko screen par show karna
plt.show()
```
