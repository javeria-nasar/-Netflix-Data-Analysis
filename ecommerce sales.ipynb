```python
import pandas as pd

# Online store ka sample sales data bana rahe hain
sales_data = {
    'OrderID': ['101', '102', '103', '104', '105', '106'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Laptop'],
    'Quantity': [1, 3, 1, 2, 1, 2],
    'Price_Per_Unit': [1200, 25, 1200, 60, 25, 1200],
    'City': ['Karachi', 'Lahore', 'Sargodha', 'Karachi', 'Sargodha', 'Lahore']
}

# Data ko table (DataFrame) mein badal rahe hain
df_sales = pd.DataFrame(sales_data)

# Har order ki Total Sales (Quantity x Price) nikal rahe hain
df_sales['Total_Sales'] = df_sales['Quantity'] * df_sales['Price_Per_Unit']

# Data dekhne ke liye
df_sales
```

```python
# Sheher ke mutabiq total sales ko jama (sum) karna
city_sales = df_sales.groupby('City')['Total_Sales'].sum()

# Result dekhne ke liye
print("City-wise Total Sales:")
print(city_sales)
```

```python
import matplotlib.pyplot as plt

# Sales ka Bar Chart banana
city_sales.plot(kind='bar', color=['darkblue', 'orange', 'purple'])

# Chart par headings aur labels lagana
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales (in USD)')

# Chart show karna
plt.show()
```
