```python
import pandas as pd

# World Population ka live working data link
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)

# Data ka pehla look dekhne ke liye
df.head()
```

```python
# Sirf Pakistan ka data filter kar rahe hain
pakistan_data = df[df['Country Name'] == 'Pakistan']

# Dekhte hain data sahi filter hua ya nahi
pakistan_data.head()
```

```python
import matplotlib.pyplot as plt

# Year aur Value (Population) ka Line Chart banana
plt.plot(pakistan_data['Year'], pakistan_data['Value'], color='green', linewidth=2.5)

# Chart par headings lagana
plt.title('Pakistan Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population (in Millions)')

# Chart screen par show karna
plt.show()
```
