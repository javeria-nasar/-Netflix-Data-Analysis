```python
import pandas as pd

# World Bank ke mutabiq China aur US ka real historical GDP growth data (in Billions USD)
data_gdp = {
    'Country Name': ['China', 'China', 'China', 'China', 'China', 'China',
                    'United States', 'United States', 'United States', 'United States', 'United States', 'United States'],
    'Year': [2000, 2005, 2010, 2015, 2020, 2024,
             2000, 2005, 2010, 2015, 2020, 2024],
    'GDP_Billion': [1211, 2285, 6087, 11060, 14680, 17790,
                    10250, 13030, 15040, 18200, 21060, 27360]
}

# Dataframe banana
df_gdp = pd.DataFrame(data_gdp)

print("Data loaded successfully!")
df_gdp.head()
```

```python
# China aur US ka data alag alag save karna
china_gdp = df_gdp[df_gdp['Country Name'] == 'China']
us_gdp = df_gdp[df_gdp['Country Name'] == 'United States']

print("China GDP Trends:")
print(china_gdp)
```

```python
import matplotlib.pyplot as plt

# Graph size aur style
plt.figure(figsize=(10, 6))

# China aur US ki growth lines draw karna
plt.plot(china_gdp['Year'], china_gdp['GDP_Billion'], color='#de2910', label='China (PRC)', linewidth=3, marker='o')
plt.plot(us_gdp['Year'], us_gdp['GDP_Billion'], color='#0a36af', label='United States', linewidth=2.5, linestyle='--', marker='s')

# Headings aur Labels
plt.title("China's Economic Rise vs United States (2000-2024)", fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP in Billions (USD)', fontsize=12)

plt.legend(fontsize=11, loc='upper left')
plt.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```
